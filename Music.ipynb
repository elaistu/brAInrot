```python
import sqlite3
import base64
import os

# List of top 5 music with their image file paths
top_5_music = [
    ["OCEAN WAVES", "static/images/OceanWaves.png"],
    ["RIVER SONG", "static/images/River Song.png"],
    ["TROPICAL VIBES", "static/images/Tropical Vibes.png"],
    ["FOREST HARMONY", "static/images/Forest Harmony.png"],
    ["MEADOW SERENADE", "static/images/Meadow Serenade.png"]
]

# Function to encode image file as base64
def encode_image_to_base64(image_filepath):
    with open(image_filepath, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')

# Connect to the SQLite database
conn = sqlite3.connect('music.db')
cursor = conn.cursor()

# Process each song title and image file path
for song_title, image_filepath in top_5_music:
    # Encode the image file as base64
    base64_image = encode_image_to_base64(image_filepath)
    
    # Update the "Thumbnail" column in the "songs" table where "Song Title" matches
    cursor.execute("""
        UPDATE songs
        SET Thumbnail = ?
        WHERE "Song Title" = ?
    """, (base64_image, song_title))

# Commit the changes and close the connection
conn.commit()
conn.close()
```
