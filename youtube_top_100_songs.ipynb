```python
from googleapiclient.discovery import build
import pandas as pd
import sqlite3

def get_youtube_service(api_key):
    return build('youtube', 'v3', developerKey=api_key)

def search_videos(youtube, query, max_results=5):
    request = youtube.search().list(
        part="snippet",
        q=query,
        type="video",
        maxResults=max_results
    )
    response = request.execute()
    return response

```

```python
def clear_db():
    conn = sqlite3.connect('youtube_song_list.db')
    c = conn.cursor()
    c.execute('DROP TABLE IF EXISTS songs')
    conn.commit()
    conn.close()
    print("Database cleared.")
```

```python
def store_in_db(songs):
    conn = sqlite3.connect('youtube_song_list.db')
    c = conn.cursor()
    c.execute('''CREATE TABLE IF NOT EXISTS songs
                 (title TEXT, video_id TEXT, description TEXT, thumbnail TEXT)''')
    c.executemany('INSERT INTO songs VALUES (?, ?, ?, ?)', songs)
    conn.commit()
    conn.close()
    print("Data stored in database.")
```

```python
api_key = 'INSERT API KEY HERE'
youtube = get_youtube_service(api_key)

# Clear the database
clear_db()

# Prompt to search for copyright-free songs
query = "trending english copyright-free music"
response = search_videos(youtube, query, max_results=100)

# Extract and store the top 5 songs
top_songs = []
for item in response['items']:
    song = (
        item['snippet']['title'],
        item['id']['videoId'],
        item['snippet']['description'],
        item['snippet']['thumbnails']['default']['url']
    )
    top_songs.append(song)
```

```python
# Store the results in the database
store_in_db(top_songs)

# Display the results
for idx, song in enumerate(top_songs, start=1):
    print(f"{idx}. {song[0]}")
    print(f"Video ID: {song[1]}")
    print(f"Description: {song[2]}")
    print(f"Thumbnail: {song[3]}")
    print(f"Watch on YouTube: https://www.youtube.com/watch?v={song[1]}")
    print()
```
